## Importando Dados

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
import locale
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

```python
dados = pd.read_csv('dadosTratados.csv', sep = ';')
```

```python
homologados = dados.query("situacao == 'Homologado'")
homologados.head()
```

```python
hom_ano = homologados.ano.value_counts()
max_ano = homologados.groupby('ano').valor_contratado_ufpr.max()

hom_max_ano = {
    'Ano' : hom_ano.index,
    'Homologações': hom_ano.values,
    'Valor Máximo Contratado': max_ano.values
}
hom_max_ano = pd.DataFrame(hom_max_ano).sort_values('Ano')
```

```python
hom_max_ano
```

```python
fig = plt.figure(figsize=(18,5), dpi=150)

fig.add_subplot(1,2,1)
ax = sns.lineplot(x='Ano', y='Homologações', data=hom_max_ano)
ax.set_title('Processos homologados por ano', loc='left', fontsize=18)
ax.set_xlabel('Ano', fontsize=14, labelpad=5)
ax.set_ylabel('Total de homologações', fontsize=14, labelpad=8)

for i in range(len(hom_max_ano)):
        ax.text(hom_max_ano.iloc[i][0] - 0.1, 
                (hom_max_ano.iloc[i][1]) - 0.5, 
                int(hom_max_ano.iloc[i][1]),
                backgroundcolor='white',
                size=9)
        
fig.add_subplot(1,2,2)
ax1 = sns.lineplot(x='Ano', y='Valor Máximo Contratado', data=hom_max_ano)
ax1.set_title('Valor máximo de homologação por ano', loc='left', fontsize=18)
ax1.set_xlabel('Ano', fontsize=14, labelpad=5)
ax1.set_ylabel('Valor Máximo', fontsize=14, labelpad=8)

for i in range(len(hom_max_ano)):
    ax1.text(hom_max_ano.iloc[i][0] - 0.4, (hom_max_ano.iloc[i][2]) - 10, 
            locale.currency(hom_max_ano.iloc[i][2], grouping=True), 
            backgroundcolor='white',
            size=9.2)
    
plt.savefig('Processos homologados e valor máximo por ano.jpeg')
```

```python
outras_modalidades = homologados.query('modalidade != "Pregão Eletrônico"')
pregao = homologados.query('modalidade == "Pregão Eletrônico"')
```

```python
modalidades = {
    'Modalidade': ['Pregão Eletrônico', 'Outras'],
    'Total': [pregao.shape[0], outras_modalidades.shape[0]],
    'Valor Máximo de Homologação': [locale.currency(pregao.valor_contratado_ufpr.max(), grouping=True), 
                                    locale.currency(outras_modalidades.valor_contratado_ufpr.max(), grouping=True)]
}

modalidades = pd.DataFrame(modalidades)
```

```python
modalidades
```

```python
fig = plt.figure(figsize=(20,5), dpi=150)

fig.add_subplot(1,2,1)
ax = sns.barplot(x='Modalidade', y='Total', data=modalidades,
                palette="tab20b")
ax.set_title('Modalidades homologadas', loc='left', fontsize=18)
ax.set_xlabel('Modalidades', fontsize=14)
ax.set_ylabel('Total', fontsize=14)

for i, v in enumerate(modalidades.Total):
    ax.text((i - 0.025), (v + 7), v)
 
fig.add_subplot(1,2,2)
ax = sns.barplot(x=outras_modalidades.modalidade.value_counts().index,
                 y=outras_modalidades.modalidade.value_counts().values,
                 data=outras_modalidades, 
                 palette="tab20b")
ax.set_title('Outras modalidades homologadas', loc='left', fontsize=18)
ax.set_xlabel('Modalidades', fontsize=14, labelpad=15)
ax.set_ylabel('Total', fontsize=14, labelpad=2)
ax.tick_params(labelrotation=10)

for i, v in enumerate(outras_modalidades.modalidade.value_counts().values):
    ax.text((i - 0.025), (v + 0.5), v)
    
plt.savefig('Modalidades e outras Homologadas.jpeg')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax = sns.barplot(x=homologados.natureza.value_counts().index,
                 y=homologados.natureza.value_counts().values,
                 data=homologados, 
                 palette="tab20b")
ax.figure.set_size_inches(13.5, 6)
ax.set_title('Natureza das licitações homologadas', loc='left', fontsize=18)
ax.set_xlabel('Natureza', fontsize=14, labelpad=7)
ax.set_ylabel('Total', fontsize=14, labelpad=2)

for i, v in enumerate(homologados.natureza.value_counts().values):
    ax.text((i - 0.075), (v + 6), v)

plt.savefig('Natureza das licitações homologadas.jpeg')
```

```python
hom_por_mes = homologados.groupby('homologacao_mes_ano').ano.count().to_frame().reset_index()
hom_por_mes.columns = ['Mês de homologação', 'Total']
hom_por_mes['Mês de homologação'] = pd.to_datetime(hom_por_mes['Mês de homologação'])
```

```python
ax = sns.lineplot(x='Mês de homologação', y='Total', data=hom_por_mes)
ax.figure.set_size_inches(12, 6)
ax.set_title('Time Series de Homologações', loc='left', fontsize=18)
ax.set_xlabel('Ano', fontsize=14, labelpad=5)
ax.set_ylabel('Homologações', fontsize=14, labelpad=8)
ax = ax
```

```python
from pandas.plotting import autocorrelation_plot

ax = plt.figure(figsize=(12,6))
ax.suptitle('Correlação do número de homologações por mês', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(hom_por_mes.Total)
ax = ax
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

resultado = seasonal_decompose(hom_por_mes.Total,  period=5)
ax = resultado.plot()
```
