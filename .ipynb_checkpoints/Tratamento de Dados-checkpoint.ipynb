# Licitações realizadas pela Universidade Federal do Paraná (2014~2020)
# Tratamento de dados

### Importando a base de dados (disponível em dados.gov)

```python
import pandas as pd
```

```python
db14 = pd.read_csv('BaseDados/financeiro_licit_2014.csv', sep = ';', encoding = 'latin-1')
db15 = pd.read_csv('BaseDados/financeiro_licit_2015.csv', sep = ';', encoding = 'latin-1')
db16 = pd.read_csv('BaseDados/financeiro_licit_2016.csv', sep = ';', encoding = 'latin-1')
db17 = pd.read_csv('BaseDados/financeiro_licit_2017.csv', sep = ';', encoding = 'latin-1')
db18 = pd.read_csv('BaseDados/financeiro_licit_2018.csv', sep = ';', encoding = 'latin-1')
db19 = pd.read_csv('BaseDados/financeiro_licit_2019.csv', sep = ';', encoding = 'latin-1')
db20 = pd.read_csv('BaseDados/financeiro_licit_2020.csv', sep = ';', encoding = 'latin-1')
```

```python
dados = pd.concat([db14, db15, db16, db17, db18, db19, db20], ignore_index = True)
dados
```

## Tratando os dados (limpeza e organização)

### Alterando tipo dos valores

```python
correcaoValor = list(dados.columns[10:14])

for v in correcaoValor:
    dados[v] = dados[v].apply(lambda x: str(x).replace('.',''))
    dados[v] = dados[v].apply(lambda x: str(x).replace(',','.'))
    dados[v] = dados[v].apply(lambda x: x.replace(x, '0') if x.isalpha() else x)
    dados[v].fillna(0)
    dados[v] = pd.to_numeric(dados[v])
dados.head(5)
```

### Retirando os valores nulos

```python
si = 'Sem Informação'
dados.fillna({'Natureza do Objeto': f'{si}', 'Modalidade de Compra': f'{si}', 'Valor Estimado UFPR': 0}, inplace = True)
```

```python
dados.info()
```

### Uniformizando coluna "Situação"

```python
sit = dados['Situação'].value_counts()
sit
```

```python
situacoes = ['Anulado', 'Suspenso', 'Fase Interna', 'Deserto', 'Homologado', 'Fracassado', 'Cancelado']
correcao = [['Anulada', 'anulada', 'anulado'], ['Suspensa', 'suspensa', 'suspenso', 'SUSPENSO'],
            ['Fase interna'], ['Deserta', 'deserto'], ['Homologada', 'homologado', 'homologada'],
            ['Fracassada'], ['cancelado na adjudicação', 'Cancelada']]

for z in range(len(situacoes)):
    dados['Situação'] = dados['Situação'].apply(lambda x: situacoes[z] if x in correcao[z] else x)
```

```python
sit = dados['Situação'].value_counts()
sit
```

### Padronização homologação

```python
# Existem processos homologados sem mês de homologação, é preciso mostrar que essa informação não está registrada

selecao = (dados['Situação'] == 'Homologado') & (dados['Homologação (mês)'].isnull())
dados[selecao] = dados[selecao].fillna({'Homologação (mês)': 'Sem Informação'})
dados[dados['Situação'] == 'Homologado'][1035:1040]
```

## Exportando dados tratados

```python
dados.to_csv('dadosTratados.csv', sep = ';', index = False)
```

## Testes de seleção

### Selecionar a frequência de cada modalidade de licitação

```python
modal = list(dados['Modalidade de Licitação'].drop_duplicates())
nt = 0
tabela = []
for i in modal:
    selecao = dados['Modalidade de Licitação'] == f'{i}'
    n1 = dados[selecao].shape[0]
    tabela.append(n1)
    nt = nt + n1

modal.append('Total de processos')
tabela.append(nt)

modalidades = pd.DataFrame(tabela, index = modal, columns = ['Quantidade'])
modalidades
```

### Selecionar a frequência de cada situação de licitação

```python
sit = list(dados['Situação'].drop_duplicates())
nt = 0
tabela = []

for i in sit:
    selecao = dados['Situação'] == f'{i}'
    n1 = dados[selecao].shape[0]
    tabela.append(n1)
    nt = nt + n1

sit.append('Total de processos')
tabela.append(nt)

situacao = pd.DataFrame(tabela, index = sit, columns = ['Quantidade'])
situacao
```

```python
dados['Diferença Contratado/Estimado'] = dados['Valor Contratado UFPR'] - dados['Valor Estimado UFPR']
dados['Vantagem'] = dados['Diferença Contratado/Estimado'] <= 0
```

```python
tabelaDiferenca = dados[dados['Situação'] == 'Homologado'][['Modalidade de Licitação','Situação','Valor Estimado UFPR', 
                                                            'Valor Contratado UFPR','Diferença Contratado/Estimado', 
                                                            'Vantagem']]
tabelaDiferenca.index = range(tabelaDiferenca.shape[0])
tabelaDiferenca.head(10)
```

```python
agrupamento = tabelaDiferenca.groupby('Modalidade de Licitação')
pd.DataFrame(agrupamento['Valor Contratado UFPR'].mean().round(2)).rename(columns = {'Valor Contratado UFPR': 'Valor médio homologado'})
```
