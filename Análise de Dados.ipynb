### Importando a base de dados (disponível em dados.gov)

```python
import pandas as pd
```

```python
db14 = pd.read_csv('BaseDados/financeiro_licit_2014.csv', sep = ';', encoding = 'latin-1')
db15 = pd.read_csv('BaseDados/financeiro_licit_2015.csv', sep = ';', encoding = 'latin-1')
db16 = pd.read_csv('BaseDados/financeiro_licit_2016.csv', sep = ';', encoding = 'latin-1')
db17 = pd.read_csv('BaseDados/financeiro_licit_2017.csv', sep = ';', encoding = 'latin-1')
db18 = pd.read_csv('BaseDados/financeiro_licit_2018.csv', sep = ';', encoding = 'latin-1')
db19 = pd.read_csv('BaseDados/financeiro_licit_2019.csv', sep = ';', encoding = 'latin-1')
db20 = pd.read_csv('BaseDados/financeiro_licit_2020.csv', sep = ';', encoding = 'latin-1')
```

```python
db14.head(5)
```

```python
dados = pd.concat([db14, db15, db16, db17, db18, db19, db20], ignore_index = True)
dados.head(5)
```

## Tratando os dados (limpeza e organização)

### Retirando os valores nulos

```python
si = 'Sem informação'
dados.fillna({'Natureza do Objeto': f'{si}', 'Modalidade de Compra': f'{si}', 'Valor Estimado UFPR': 0}, inplace = True)
```

```python
dados.info()
```

### Uniformizando coluna "Situação"

```python
sit = list(dados['Situação'].drop_duplicates())
sit
```

```python
correcao_anulado = ['Anulada', 'anulada', 'anulado']
correcao_suspenso = ['Suspensa', 'suspensa', 'suspenso', 'SUSPENSO']
correcao_faseinterna = ['Fase interna']
correcao_deserto = ['Deserta', 'deserto']
correcao_homologado = ['Homologada', 'homologado', 'homologada']
correcao_fracassado = ['Fracassada']
correcao_cancelado = ['cancelado na adjudicação', 'Cancelada']

dados['Situação'] = dados['Situação'].apply(lambda x: 'Anulado' if x in correcao_anulado else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Suspenso' if x in correcao_suspenso else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Fase Interna' if x in correcao_faseinterna else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Deserto' if x in correcao_deserto else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Homologado' if x in correcao_homologado else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Fracassado' if x in correcao_fracassado else x)
dados['Situação'] = dados['Situação'].apply(lambda x: 'Cancelado' if x in correcao_cancelado else x)
```

```python
sit = list(dados['Situação'].drop_duplicates())
sit
```

### Padronização homologação

```python
# Existem processos homologados sem mês de homologação, é preciso mostrar que essa informação não está registrada
selecao = (dados['Situação'] == 'Homologado') & (dados['Homologação (mês)'].isnull())
dados[selecao] = dados[selecao].fillna({'Homologação (mês)': 'Sem informação'})
dados[dados['Situação'] == 'Homologado'][1035:1040]
```

## Testes de seleção

### Selecionar a frequência de cada modalidade de licitação

```python
modal = list(dados['Modalidade de Licitação'].drop_duplicates())
nt = 0

print('-> Números de processos licitatórios por modalidade \n')
for i in modal:
    selecao = dados['Modalidade de Licitação'] == f'{i}'
    n1 = dados[selecao].shape[0]
    print(f'{i}: {n1}')
    nt = nt + n1
print (f'\nTotal de processos: {nt}')
```

### Selecionar a frequência de cada situação de licitação

```python
print('-> Números de processos licitatórios por situação \n')
nt = 0

for i in sit:
    selecao = dados['Situação'] == f'{i}'
    n1 = dados[selecao].shape[0]
    print(f'{i}: {n1}')
    nt = nt + n1
print (f'\nTotal de processos: {nt}')
```
